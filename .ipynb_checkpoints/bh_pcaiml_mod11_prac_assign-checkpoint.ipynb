### Data Preparation
#### Approach to handling missing values

<ul>
	<li>Small missing values: Drop rows.</li>
	<li>Large missing values in a column: Drop the column.</li>
    <li>Numerical features: Mean/median/mode or interpolation.</li>
    <li>Categorical features: Mode or "Unknown" category.</li>
</ul>

```python
# VIN column - Vehicle Identification Number encodes lot of car features however since we have separate columns, 
# for most of the VIN features, we dont need to decode VIN. We can drop the VIN column. 
# ID Column - ID, Like wise drop the ID column as it does not capture any feature of the vehicle
# SIZE column - a Large portion of dataset (306k out of 426K) rows have no size defined. Hence dropping it. 
# Region Column - While the regions may give more granular insights, there are 404 regions across 51 states. 
# Dropping region for now to keep it simple
df = df.drop(['id', 'VIN','size',"region"], axis=1)

#remove all the wild characters from model column
#df['model'] = df['model'].str.replace("'","").str.replace('"','').str.replace("&","").str.replace("$","").str.replace("â™¿","").str.replace("$","").str.replace("(","").str.replace("/","").str.replace("%","").replace("*","")
#df['model'] = df['model'].str.replace("'","_").str.replace('"','_')

# Though the model column could give good insights, there are a total of ~29 thousand unique models. 
# It becomes significantly difficult to process the model column. Dropping this column to keep it simple. 
df = df.drop(["model"], axis=1)

# Replacing the missing values
df.fillna({"fuel":'gas'}, inplace=True)
df.fillna({"title_status":'missing'}, inplace=True)
df.fillna({"drive":'unknown'}, inplace=True)
df.fillna({"paint_color":'unknown'}, inplace=True)
df.fillna({"condition":'unknown'}, inplace=True)
df.fillna({"type":'unknown'}, inplace=True)
df.fillna({"size":'unknown'}, inplace=True)
df.fillna({"model":'unknown'}, inplace=True)
df.fillna({"manufacturer":'unknown'}, inplace=True)

# There are 1025 rows with missing year
# There are 4400 rows with missing odometer 
# The number of rows is less than 1% of the entire data set for of 426K rows. 
# Dropping those rows.
df.dropna(subset=['year','odometer'], inplace=True)

# The value entries in the year column have DOT at the end. 
# removing it and converting the column to integer
df['year'] = df['year'].astype(str).str.rstrip('.')
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')

## convert odometer to integer 
df['odometer'] = df['odometer'].astype(int)

# setting "manual" as the tranmission type when tranmission is missing. 
df.fillna({"transmission":'manual'}, inplace=True)

print(f"Cylinders mode: {df['cylinders'].mode()[0]}")

df['cylinders'].fillna(df['cylinders'].mode()[0], inplace=True)
```

```python
# Define quantile thresholds
q_low = df["price"].quantile(0.05)  # 5th percentile
q_hi  = df["price"].quantile(0.95)  # 95th percentile

# Find rows that fall outside the range
outliers = df[(df["price"] < q_low) | (df["price"] > q_hi)]

low_outliers = df[(df["price"] < q_low) ]
high_outliers = df[(df["price"] > q_hi)]

clean_df = df[(df["price"] > q_low) & (df["price"] < q_hi)]

# Print number of outliers
print(f"Low Outlier count: {low_outliers.shape[0]}")
print(f"High Outlier count: {high_outliers.shape[0]}")

print(f"Number of rows outside the range: {outliers.shape[0]}")

print(f"Mean:{df['price'].mean()},Median:{df['price'].median()}, Mode:{df['price'].mode()[0]}, SD:{df['price'].std()}")
print(f"Mean:{clean_df['price'].mean()},Median:{clean_df['price'].median()}, Mode:{clean_df['price'].mode()[0]}, SD:{clean_df['price'].std()}")
```

By analyzing the count plot of categorical features, we can identify two types of categorical data: ordinal and nominal.

Nominal Data: These categories have no intrinsic order (e.g., colors, car brands). To handle them in machine learning models, we typically apply one-hot encoding, which creates binary columns for each category.
Ordinal Data: These categories follow a meaningful order (e.g., education level, ranking). Instead of one-hot encoding, we should map them to numerical values (e.g., "low" â†’ 1, "medium" â†’ 2, "high" â†’ 3) to preserve their inherent order.
This distinction ensures that models correctly interpret categorical data, preventing loss of meaningful relationships in ordinal variables.

# Modeling

## Data Preprocessing

From our exploratory data analysis, we found that almost all variables are categorical. Some of these can be combined to create a more meaningful new feature. To keep things simple, we will encode the remaining categorical features using OneHotEncoder, as we are primarily using sklearn for model building.

However, one drawback of OneHotEncoding is that it can negatively impact the performance of decision trees and random forests. Despite this, we must use it because RandomForestClassifier in sklearn only supports numerical data.

Considerations for Feature Engineering:
Should we apply dimensionality reduction?
After OneHotEncoding, the total number of features may exceed 50, and some frequency-related columns may be highly correlated. This makes dimensionality reduction techniques like PCA a potential option to simplify the data while preserving important information.

Should we apply feature expansion?
Creating new meaningful features from categorical data manually is challenging. However, we can look at methods like the Kernel SVM, which demonstrates the power of feature expansion. In our case, we predict that the Gaussian RBF kernel will work better than the polynomial kernel. Since most variables are categorical, measuring similarity (as the RBF kernel does) might be more useful than applying polynomial transformations.

```python
X = clean_df.drop(columns=['price'])
y = clean_df['price']

print(X.shape)
print(y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
num_features = X_train.select_dtypes(['int64', 'float64']).columns
cat_features = X_train.select_dtypes(['object']).columns

full_pipeline = ColumnTransformer([('num', StandardScaler(), num_features),
                                   ('cat', OneHotEncoder(), cat_features)])

X_train = full_pipeline.fit_transform(X_train)
X_test = full_pipeline.transform(X_test)

# Extract new feature names
num_feature_names = num_features  # Numeric columns remain unchanged
cat_feature_names = full_pipeline.named_transformers_['cat'].get_feature_names_out(cat_features)

# Combine all feature names
new_feature_names = list(num_feature_names) + list(cat_feature_names)

#print numeric features
print(num_feature_names)

#print categorical features 
print(cat_feature_names)

# Print the new feature names
print(new_feature_names)
```

### Ridge Regression

```python
params_dict = {'alpha': [0.1, 1.0, 10.0]}
ridge = Ridge()

logreg_gscv = GridSearchCV(ridge, 
                           param_grid=params_dict)

logreg_gscv.fit(X_train, y_train)

evaluate_model(logreg_gscv)
```

```python
results = permutation_importance(logreg_gscv, X_test.toarray(), y_test, n_repeats=10, random_state=42)

importance = results.importances_mean

clean_names = [i.split('__')[-1] for i in new_feature_names]

print('Features (in order of importance):')

rows_list = []
for feature, importance in sorted(zip(clean_names, importance), key=lambda x: x[1], reverse=True):
    if (f'{importance:.2f}' == "0.00"):
        continue
    rows_list.append([feature,importance]) 
    print(f'{feature}: {importance:.5f}')

feature_importance_df = pd.DataFrame(rows_list,columns=('features','importance')) 

feature_importance_df.plot.barh(y="importance", x= "features", figsize=(9, 5), color="green")

plt.title("Ridge model with regularization")
plt.axvline(x=0, color=".5")
plt.xlabel("coefficient values")
plt.subplots_adjust(left=0.3)

plt.savefig('images/feature_importance.png')
```

```python
manufacturer_names = [i.split('__')[-1].replace("manufacturer_","") for i in new_feature_names]

coef_df = pd.DataFrame({'feature': manufacturer_names, 'coefs': logreg_gscv.best_estimator_.coef_})
coef_df = coef_df.sort_values(by = 'coefs', ascending = False)

top_and_bottom_coef_df = coef_df.query("coefs >= 5000 or coefs <= -4000")[["feature","coefs"]]
top_and_bottom_coef_df
```
